```python
members_list = [['1910800', 'Mostafa Hasan', 2000.0, 0.0, 'paid'],
                ['2111220', 'Mahabubul Alam', 2000.0, 0.0, 'paid'],
                ['2110315', 'Mir Aurisha Mehek', 2000.0, 0.0, 'paid'],
                ['2110806', 'Sazin Bin Noor', 0.0, 2000.0, 'unpaid'],
                ['2110969', 'Md Nazmul Hasan', 0.0, 2000.0, 'unpaid'],
                ['2120764', 'Sauda Khan', 0.0, 2000.0, 'unpaid'],
                ['2120263', 'Arvil Nath', 0.0, 2000.0, 'unpaid'],
                ['2111584', 'Al Mahmud Shafin', 0.0, 2000.0, 'unpaid'],
                ['2110686', 'Sadia Afrin', 0.0, 2000.0, 'unpaid'],
                ['2022766', 'Snigdha Oishi', 0.0, 2000.0, 'unpaid']]
```

```python
#Define function for displaying the Members List
def print_list():
  print("Members List:")
  print("-----------")
  print("Membership Number   Name\t\t   Cash Paid\t   Cash Due\tStatus")
  for i in range(0, len(members_list)):
    print("{0:18}  {1:18}  {2:9}\t{3:9}  \t{4:7}".format(members_list[i][0], members_list[i][1], members_list[i][2], members_list[i][3],members_list[i][4]))
```

```python
#Define function for inserting a new member
def register():
  #Ask for user detail
  print("****** Registration ******")
  print("\n") #For new line
  membership_number = input("Enter Student ID: ")
  name = input("Enter Name: ")
  cash_paid = float(input("Enter Paid Amount: "))
  cash_due = 2000 - cash_paid
  if(cash_due == 0):
    status = "paid"
  else:
    status = "unpaid"
  #Adding new user
  New_user = [membership_number, name, cash_paid, cash_due, status]
  members_list.append(New_user)
  print("\n") #For new line
  print("Registration Complete. \nThank You for Choosing us.")
```

```python
#Define function for Checking total fees collected
def total_fees_collected():
  print("****** Total Fees Collected ******")
  print("\n") #For new line
  print_list()
  print("\n") #For new line
  total = 0
  for i in range(0, len(members_list)):
    collected_fees = members_list[i][2]   
    total = total + collected_fees
  paid_status_count = sum([i.count('paid') for i in members_list]) #count of members whose payment status is paid
  print("Total Number of Members: ",len(members_list))
  print("Total Payment Completed Members: ",paid_status_count)
  print("Total Fees Collected: ",total)
```

```python
#Define function for Collecting fees Tk.2000/- ,updating and printing receipt
def receipt(edit_ID,new_cash_paid):
  print("****** Receipt ******")
  print("Received From: ",members_list[edit_ID][1])
  print("Paid Amount: ",new_cash_paid)
  print("Total Paid Amount: ",members_list[edit_ID][2])
  print("Due: ",members_list[edit_ID][3])
  print("Payment Status: ",members_list[edit_ID][4])
  print("\n") #For new line

def fees_collection():
  print("****** Fees Collection ******")
  print("\n") #For new line
  print_list()
  print("\n") #For new line
  membership_number = input("Enter Student ID: ")
  print("\n") #For new line
  found = 1
  for i in range ( 0, len(members_list)):
    if (members_list[i][0]) == membership_number :
      for membership_number in members_list:
        edit_ID = members_list.index(membership_number)   #index number of the member list
      new_cash_paid = float(input("Enter Paid Amount: "))
      previously_paid = members_list[i][2]
      cash_paid = previously_paid + new_cash_paid
      previously_due = members_list[i][3]
      cash_due = previously_due - new_cash_paid
      if(cash_due == 0):
        status = "paid"
        members_list[edit_ID][2] = cash_paid
        members_list[edit_ID][3] = cash_due
        members_list[edit_ID][4] = status
        print("\n") #For new line
        receipt(edit_ID,new_cash_paid)
        found = 0
      else:
        status = "unpaid"
        members_list[edit_ID][2] = cash_paid
        members_list[edit_ID][3] = cash_due
        members_list[edit_ID][4] = status
        print("\n") #For new line
        receipt(edit_ID,new_cash_paid)
        found = 0
  if found == 1:  
    print('Member Not Found.')
```

```python
#Define function for Printing List of persons whose fees is not paid
def fees_unpaid_list():
  print("****** Unpaid Members List ******")
  print("\n") #For new line
  print("Membership Number   Name\t\t   Cash Paid\t   Cash Due\tStatus")
  found = 1
  for i in range ( 0, len(members_list)):
    if (members_list[i][4]) == 'unpaid' :     
      print("{0:18}  {1:18}  {2:9}\t{3:9}  \t{4:7}".format(members_list[i][0], members_list[i][1], members_list[i][2], members_list[i][3],members_list[i][4]))
      found = 0
  if found == 1:
    print("\n") #For new line
    print("Everyone Paid Their Fees.")
  unpaid_status_count = sum([i.count('unpaid') for i in members_list])
  print("\n") #For new line
  print("Total Fees Unpaid Members: ",unpaid_status_count)
```

```python
option = 10 #Arbitrary value for entering while loop

print("****** IUB Students' Club ******")
print("\n") #For new line
#Loop to repeatedly display the menu
while option != 0: #Exit loop when the value is 0
  print("What do you wish to do?")
  print("1 -> Registration")
  print("2 -> Print the Members List")
  print("3 -> Check Total Fees Collected")
  print("4 -> Fees Collect and Print Receipt")
  print("5 -> Fees Unpaid Members List")
  print("0 -> Exit Menu")
  print("\n") #For new line4
  option = int(input("Select option: "))
  print("\n") #For new line
 
  #Check chosen option and call the appropriate function
  if option == 0:
    print("Thank you for using the system.") 
  elif option == 1:
    #Call the register function
    register()
    print("\n") #For new line
  elif option == 2:
    #Call the print_list function
    print_list()
    print("\n") #For new line
  elif option == 3:
    #Call the total_fees_collected function
    total_fees_collected()
    print("\n") #For new line
  elif option == 4:
    #Call the fees_collection function
    fees_collection()
    print("\n") #For new line
  elif option == 5:
    #Call the fees_unpaid_list function
    fees_unpaid_list()
    print("\n") #For new line
  else:
    print("Non-existing Option.")
    print("\n") #For new line
```
